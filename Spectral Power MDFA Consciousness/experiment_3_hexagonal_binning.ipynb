```r
# Yacine Mahdid 2020-01-10
# The goal of this script is to generate a hexagonal binning plot using ggplot2
# in order to better visualize the relationship in a dense point cloud.
install.packages("ggplot2")
library("ggplot2")
```

```r

# We need to load the data as a df
df <- read.table("motif_power.csv", 
                 header = TRUE,
                 sep = ",")

R = cor(df$motif_frequency, df$power)
# Color palette for the correlation
ggplot(df,(aes(x=motif_frequency, y=power))) + 
    stat_binhex(colour="white") + 
    labs(x="Motif Frequency", y="Power") +
    ggtitle(sprintf("Power vs Motif 7 Frequency at Alpha (R = %.2f)",R)) + 
    theme(plot.title = element_text(hjust = 0.5))
```
